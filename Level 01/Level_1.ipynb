```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\\Bootcamp\\Phase 1\\Level 10\\dataset\\Mall_Customers.csv", encoding="ISO-8859-1")

print(df.head())

mean_value = df["Annual Income (k$)"].mean()
median_value = df["Annual Income (k$)"].median()
mode_value = df["Annual Income (k$)"].mode()[0]

df["Genre"] = pd.cut(df["Annual Income (k$)"],
                                 bins=[0, 10, 50, df["Annual Income (k$)"].max()], 
                                 labels=["Low", "Medium", "High"])

customer_segments = df["Genre"].value_counts()

print(f"Mean: {mean_value}, Median: {median_value}, Mode: {mode_value}")

print(customer_segments)

plt.figure(figsize=(8, 5))
sns.barplot(x=customer_segments.index, y=customer_segments.values, palette="coolwarm")
plt.xlabel("Spending Category")
plt.ylabel("Number of Customers")
plt.title("Customer Segmentation Based on Spending")
plt.show()


```

```python
#Level 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\\Bootcamp\\Phase 1\\Level 10\\dataset\\data-2.csv")
print(df.head())

price_col = "UnitPrice"

df = df[[price_col]].dropna()

print(df.describe())

price_range = df[price_col].max() - df[price_col].min()
price_variance = np.var(df[price_col], ddof=1)  
price_std_dev = np.std(df[price_col], ddof=1)

print(f"Price Range: {price_range}")
print(f"Price Variance: {price_variance:.2f}")
print(f"Price Standard Deviation: {price_std_dev:.2f}")

sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))
sns.histplot(df[price_col], bins=30, kde=True, color="blue")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Product Price Distribution")
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x=df[price_col], color="orange")
plt.xlabel("Price")
plt.title("Boxplot of Product Prices")
plt.show()
```

```python
#Level 3
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\\archive\\Details.csv")

print(df.head())

print(df.columns)

region_1 = "East"  
region_2 = "West"

group_1 = df[df["Category"] == region_1]["Amount"]
group_2 = df[df["Category"] == region_2]["Amount"]

print(f"Group {region_1} size:", len(group_1))
print(f"Group {region_2} size:", len(group_2))

t_stat, p_value = stats.ttest_ind(group_1, group_2, equal_var=False)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. Sales differ significantly between regions.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference in sales.")

sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))
sns.boxplot(x="Category", y="Amount", data=df, palette="coolwarm")

plt.xlabel("Region")
plt.ylabel("Sales")
plt.title("Sales Distribution by Region")
plt.show()
```

```python
#Level 4
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

# Load the dataset
df = pd.read_csv("D:\\archive (1)\\housing.csv")

# Display first few rows
print(df.head())

# Display column names
print(df.columns)
feature_1 = "total_rooms"  # Independent variable (e.g., square footage)
feature_2 = "median_income"  # Dependent variable (e.g., house price)

# Check for missing values and drop them if necessary
df = df[[feature_1, feature_2]].dropna()
pearson_corr, pearson_p = pearsonr(df[feature_1], df[feature_2])
print(f"Pearson Correlation: {pearson_corr:.3f} (p-value: {pearson_p:.3f})")
spearman_corr, spearman_p = spearmanr(df[feature_1], df[feature_2])
print(f"Spearman Correlation: {spearman_corr:.3f} (p-value: {spearman_p:.3f})")
# Set seaborn style
sns.set_style("whitegrid")

# Scatter plot with regression line
plt.figure(figsize=(8, 5))
sns.regplot(x=df[feature_1], y=df[feature_2], scatter_kws={"color": "blue"}, line_kws={"color": "red"})

# Labels and title
plt.xlabel(feature_1)
plt.ylabel(feature_2)
plt.title(f"Scatter Plot: {feature_1} vs. {feature_2}")
plt.show()
```
