```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=200, centers=3, cluster_std=1.5, random_state=42)
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

linkage_matrix = linkage(df_scaled, method='ward')

plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

num_clusters = 3
cluster_labels = fcluster(linkage_matrix, num_clusters, criterion='maxclust')

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Feature1'], y=df['Feature2'], hue=cluster_labels, palette="viridis")
plt.title("Hierarchical Clustering (3 Clusters)")
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.legend(title="Cluster")
plt.show()

print("The dendrogram helps in choosing the number of clusters by looking for large vertical gaps in the hierarchy. \n"
      "Here, we chose 3 clusters based on a noticeable separation in the linkage structure.")
```
