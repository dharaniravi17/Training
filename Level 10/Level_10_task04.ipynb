```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr
np.random.seed(42)
size = np.random.randint(500, 5000, 200)
price = size * np.random.uniform(100, 300) + np.random.normal(5000, 20000, 200)

df = pd.DataFrame({'Size': size, 'Price': price})
pearson_corr, pearson_p = pearsonr(df['Size'], df['Price'])
spearman_corr, spearman_p = spearmanr(df['Size'], df['Price'])
print(f"Pearson Correlation: {pearson_corr:.4f} (p-value: {pearson_p:.4f})")
print(f"Spearman Correlation: {spearman_corr:.4f} (p-value: {spearman_p:.4f})")

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Size'], y=df['Price'], alpha=0.7)
plt.title(f"Housing Prices vs. Size\nPearson Corr: {pearson_corr:.2f}")
plt.xlabel("Size (sqft)")
plt.ylabel("Price ($)")
plt.grid(True)
plt.show()
```
