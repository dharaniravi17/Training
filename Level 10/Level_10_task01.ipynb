```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\\Bootcamp\\Phase 1\\Level 10\\dataset\\Mall_Customers.csv", encoding="ISO-8859-1")

print(df.head())

mean_value = df["Annual Income (k$)"].mean()
median_value = df["Annual Income (k$)"].median()
mode_value = df["Annual Income (k$)"].mode()[0]

df["Genre"] = pd.cut(df["Annual Income (k$)"],
                                 bins=[0, 10, 50, df["Annual Income (k$)"].max()], 
                                 labels=["Low", "Medium", "High"])

customer_segments = df["Genre"].value_counts()

print(f"Mean: {mean_value}, Median: {median_value}, Mode: {mode_value}")

print(customer_segments)

plt.figure(figsize=(8, 5))
sns.barplot(x=customer_segments.index, y=customer_segments.values, palette="coolwarm")
plt.xlabel("Spending Category")
plt.ylabel("Number of Customers")
plt.title("Customer Segmentation Based on Spending")
plt.show()


```
